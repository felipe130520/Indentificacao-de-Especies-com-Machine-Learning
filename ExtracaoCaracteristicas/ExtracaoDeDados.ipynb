```python
##Importando Bibliotecas
import os, io, time
import pandas as pd
import numpy as np
import warnings
from skimage.io import imread, imshow
from skimage.transform import resize
from skimage.color import rgb2gray
from skimage.feature import hog
from skimage.feature import local_binary_pattern
from skimage import img_as_ubyte
```

```python
##Carregando imagens

os.chdir('./Imagens')

arquivosImagens = [f for f in os.listdir() if f.endswith('.jpg')]
```

```python
##Extraindo dados com o HOG

resolucoes = [(128, 128), (256, 256)]
pixelsPorCelula = [(32, 32), (16, 16), (8, 8)]


os.chdir('../Bases')
os.listdir()


for res in resolucoes:
    for ppc in pixelsPorCelula:
        hogFeatures = []

        os.chdir('../Imagens')
        for filename in arquivosImagens:
            try:
                # 1. Carrega em tons de cinza
                imageGray = imread(filename, as_gray=True)
                
                # 2. Faz o redimensionamento
                imageResized = resize(imageGray, res, anti_aliasing=True)
                
                # 3. Extrai as características do HOG
                fd, hogImage = hog(imageResized, orientations=9, pixels_per_cell=ppc,
                                   cells_per_block=(2, 2), visualize=True)
                hogFeatures.append(fd)
                
            except ValueError as e:
                # Se o erro acontecer, captura e exibe o culpado sem parar o código
                print(f"\n[ALERTA] Erro ao processar o arquivo: {filename}")
                print(f"-> Formato original carregado: {imageGray.shape if 'imageGray' in locals() else 'Não carregou'}")
                print(f"-> Formato após resize: {imageResized.shape if 'imageResized' in locals() else 'Não redimensionou'}")
                print(f"-> Mensagem do erro: {e}\n")
                
                # Opcional: Se quiser pular essa imagem na lista de classes depois,
                # pode ser necessário remover o nome dela de 'arquivosImagens' ou criar uma lista limpa.
                continue
        os.chdir('../Bases')

        # Substituindo as raças de cachorros por 'dog'(para cachorros) e 'cat'(para gatos)
        classe = ['dog' if 'american_bulldog' in file.lower()
                  or 'american_pit_bull_terrier' in file.lower()
                  or 'basset_hound' in file.lower()
                  or 'beagle' in file.lower()
                  or 'boxer' in file.lower()
                  or 'chihuahua' in file.lower()
                  or 'english_cocker_spaniel' in file.lower()
                  or 'english_setter' in file.lower()
                  or 'german_shorthaired' in file.lower()
                  or 'great_pyrenees' in file.lower()
                  or 'havanese' in file.lower()
                  or 'japanese_chin' in file.lower()
                  else 'cat' if 'abyssinian' in file.lower()
                  or 'bengal' in file.lower()
                  or 'birman' in file.lower()
                  or 'bombay' in file.lower()
                  or 'british_shorthair' in file.lower()
                  or 'egyptian_mau' in file.lower()
                  or 'maine_coon' in file.lower()
                  or 'persian' in file.lower()
                  or 'ragdoll' in file.lower()
                  or 'russian_blue' in file.lower()
                  or 'siamese' in file.lower()
                  or 'sphynx' in file.lower() else file for file in arquivosImagens]
        
        dfFeatures = pd.DataFrame(hogFeatures)
        dfFeatures.columns = [f'hog_feature_{i+1}' for i in range(dfFeatures.shape[1])]
        dfFeatures['classe'] = classe

        filenameCSV = f'HOG_{res[0]}_{ppc[0]}x{ppc[1]}.csv'
        dfFeatures.to_csv(filenameCSV, sep=',', index=False)

        print(f'Arquivo {filenameCSV} salvo com suscesso! Colunas : {dfFeatures.shape[1]}')

print('\n--- Todos os processos comcluidos com suscesso! ---')
```

```python
##Extraindo dados com o LBP


os.chdir('../Bases')
os.listdir()


listaRadius = [12, 6, 3]
resolucoes = [(128, 128), (256, 256)]

  


for res in resolucoes:
    for rad in listaRadius:
        lbpFeatures = []
        n_points = 8 * rad

        os.chdir('../imagens')
        for filename in arquivosImagens:
            image = imread(filename, as_gray=True)
            imageResized = resize(image, res, anti_aliasing=True)

            imageResized = img_as_ubyte(imageResized)

            lbp = local_binary_pattern(imageResized, n_points, rad, method="uniform")

            (hist, _) = np.histogram(lbp.ravel(), bins=np.arange(0, n_points + 3), range=(0, n_points+2))

            hist = hist.astype("float")
            hist /= (hist.sum() +1e-7)
            lbpFeatures.append(hist)

        os.chdir('../Bases')

        # Substituindo as raças de cachorros por 'dog'(para cachorros) e 'cat'(para gatos)
        classe = ['dog' if 'american_bulldog' in file.lower()
                  or 'american_pit_bull_terrier' in file.lower()
                  or 'basset_hound' in file.lower()
                  or 'beagle' in file.lower()
                  or 'boxer' in file.lower()
                  or 'chihuahua' in file.lower()
                  or 'english_cocker_spaniel' in file.lower()
                  or 'english_setter' in file.lower()
                  or 'german_shorthaired' in file.lower()
                  or 'great_pyrenees' in file.lower()
                  or 'havanese' in file.lower()
                  or 'japanese_chin' in file.lower()
                  else 'cat' if 'abyssinian' in file.lower()
                  or 'bengal' in file.lower()
                  or 'birman' in file.lower()
                  or 'bombay' in file.lower()
                  or 'british_shorthair' in file.lower()
                  or 'egyptian_mau' in file.lower()
                  or 'maine_coon' in file.lower()
                  or 'persian' in file.lower()
                  or 'ragdoll' in file.lower()
                  or 'russian_blue' in file.lower()
                  or 'siamese' in file.lower()
                  or 'sphynx' in file.lower() else file for file in arquivosImagens]
        
        dfLBP = pd.DataFrame(lbpFeatures)
        dfLBP.columns = [f'lbp_feature_{i+1}' for i in range(dfLBP.shape[1])]

        dfLBP['classe'] = classe

        filenameCSV = f'LBP_{res[0]}_{rad}r.csv'
        dfLBP.to_csv(filenameCSV, sep=',', index=False)
        
        print(f'Arquivo {filenameCSV} salvo com suscesso! Colunas : {dfLBP.shape[1]}')

print('\n--- Todos os processos foram concluidos! ---')
        
```
