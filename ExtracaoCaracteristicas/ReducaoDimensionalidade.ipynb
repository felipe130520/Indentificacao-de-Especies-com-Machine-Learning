```python
#Importando bibliotecas
import os, io, time
import pandas as pd
import glob
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

os.chdir('./bases')
os.listdir()
```

```python
##Lê as bases

arquivosCSV = [f for f in os.listdir() if f.endswith('.csv')]

bases = {}

for arquivo in arquivosCSV:
    bases[arquivo] = pd.read_csv(arquivo)
    print(f"arquivo {arquivo} carregado com suscesso")
```

```python
##Redução de Dimensionalidade com o PCA

# Lista de todos os seus arquivos gerados anteriormente
# O glob.glob pega todos os arquivos que começam com HOG_ ou LBP_ e terminam em .csv
variancias = [0.75, 0.90]

for arquivo in bases:
    print(f"\n--- Processando {arquivo} ---")

    # 1. Carregar os dados
    df = pd.read_csv(arquivo)

    # Separar a coluna de classe (que não entra no PCA) das features
    # Assumindo que a coluna de classe chama-se 'label' ou 'classe'
    colunaClasse = 'label' if 'label' in df.columns else 'classe'

    X = df.drop(columns=[colunaClasse])
    y = df[colunaClasse]

    # 2. Normalização (CRUCIAL para o PCA funcionar corretamente)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    for var in variancias:
        # 3. Aplicar PCA
        # n_components como float seleciona o número de eixos para manter a variância desejada
        pca = PCA(n_components=var)
        X_pca = pca.fit_transform(X_scaled)

        # Criar DataFrame com as novas componentes
        dfPCA = pd.DataFrame(X_pca)
        dfPCA.columns = [f'pca_component_{i+1}' for i in range(dfPCA.shape[1])]

        # Readicionar a classe ao final
        dfPCA[colunaClasse] = y.values

        # 4. Salvar o novo dataset
        # Ex: HOG_128_16x16_PCA75.csv
        nomeSaida = arquivo.replace(".csv", f"_PCA{int(var*100)}.csv")
        dfPCA.to_csv(nomeSaida, index=False)

        print(f"Variação {var*100}%: Reduzido de {X.shape[1]} para {X_pca.shape[1]} colunas. Salvo como {nomeSaida}")

print("\n--- Redução de dimensionalidade concluída! ---")
```
