```python
##Importando bibliotecas
import os, io, time
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import KFold

```

```python
##Acessando bases geradas com os extratores e a redução de dimensionalidade
os.chdir('../ExtracaoCaracteristicas/Bases')
os.listdir()

arquivosCSV = [f for f in os.listdir() if f.endswith('.csv')]

bases = {}

for arquivo in arquivosCSV:
    bases[arquivo] = pd.read_csv(arquivo)
    print(f"Arquivo {arquivo} carregado com suscesso")
```

```python
# Rode isso para testar uma das bases
nome_teste = list(bases.keys())[0]
print(bases[nome_teste]['classe'].unique())
```

```python
##KNN com holdout 70/30


for nomeArquivo, dataset in bases.items():
  X = dataset.iloc[:, :-1].values

  y = dataset.iloc[:, -1].values

  X_train_70, X_test_30, y_train_70, y_test_30 = train_test_split(X, y, test_size=0.3, random_state= 23 )

  for i in range(1, 11):
    knn = KNeighborsClassifier(n_neighbors= i , metric='euclidean')

    knn.fit(X_train_70, y_train_70)

    y_pred = knn.predict(X_test_30)

    print(f"{nomeArquivo}:Knn{i}:{f1_score(y_test_30, y_pred, average='weighted'):.4f}")
```

```python
##KNN com kfold de 10

for nomeArquivo, dataset in bases.items():
  X = dataset.iloc[:, :-1].values

  y = dataset.iloc[:, -1].values

  kf = KFold(n_splits=10, random_state=1, shuffle=True)

  for i in range(1, 11):
    knn = KNeighborsClassifier(n_neighbors= i , metric='euclidean')

    scores = cross_val_score(knn, X, y, scoring='accuracy', cv=kf)

    y_pred = cross_val_predict(knn, X, y, cv=kf)

    print(f"{nomeArquivo}:Knn{i}:{f1_score(y, y_pred, average="weighted"):.4f}")
```
